# Expressing as linear combination

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

fig, ax = plt.subplots()

Origin = [0,0]
u1=[2,-3]
u2=[6,4]
x=[9,-7]


if np.dot(u1,u2)==0:
    c1 = (np.dot(x,u1)/np.dot(u1,u1))
    c2 = (np.dot(x,u2)/np.dot(u2,u2))
    p=np.array(u1)*c1+np.array(u2)*c2
    print("x=",p)
    
    print('Projection of x onto u: ',p)
    print('Orthogonal Complement of p: ',x-p)


    ax.quiver(*Origin,*u1,color = 'red',scale=10)
    ax.quiver(*Origin,*u2,color = 'red',scale=10)
    ax.quiver(*Origin,*x,color = 'blue',scale=10)



    ax.quiver(*Origin,*p,color = 'green',scale=10)


    ax.quiver(*Origin,*(x-p),color='green',scale=10)


    ax.axis([-10,10,-10,10])
    plt.show()
else:
    print("Not Orthogonal")
```

```python
import matplotlib.pyplot as plt

%matplotlib notebook

from mpl_toolkits.mplot3d import axes3d
import numpy as np

#add %matplotlib inline if using a Jupyter notebook, remove if using a .py script
%matplotlib inline


origin = [0,0,0]
u1=[2,5,-1]
u2=[-2,1,1]
y=[1,2,3]

c1=(np.dot(y,u1)/np.dot(u1,u1))
c2=(np.dot(y,u2)/np.dot(u2,u2))

p=np.array(u1)*c1+np.array(u2)*c2

print('Projection of y onto u: ',p)
print('Orthogonal Complement of p: ',y-p)

fig=plt.figure()

ax=fig.gca(projection='3d')

ax.quiver(*origin, *u1, length=0.1,color=["red"], normalize=True)
ax.quiver(*origin, *u2, length=0.1,color=["red"], normalize=True)
ax.quiver(*origin, *y, length=0.1,color=["blue"], normalize=True)

ax.quiver(*origin, *p, length=0.1,color=["green"], normalize=True)

ax.quiver(*origin, *(y-p), length=0.1,color=["green"], normalize=True)

plt.show()
```
